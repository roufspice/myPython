기생충에 참여한 회사들의 리스트를 정리해보자! 

+ 2019년도 영화목록을 DataFrame 형식으로 가져온다
+ 그 중 '기생충'에 관련한 영화정보를 얻는다.
+ '기생충'에 참여한 업체들의 리스트를 가져온다

```python
movie19_URL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=ae65dcfa77374933afc5fd054760188c&itemPerPage=50&openStartDt=2019&openEndDt=2019"
#2019년 기준 개봉된 영화 50편을 임시로 가져옴! 
moviePage = urllib.request.urlopen(movie19_URL)

#json파일형식으로 변형해준다.
moviePage_json = json.loads(moviePage.read().decode('utf-8'), encoding='utf-8')

#type(moviePage.read()) => byte형태로 불러옴
#moviePage.read().decode('utf-8')=> byte형태의 데이터를 utf-8형식의 유니코드로 디코딩화 => 데이터타입: 문자열
#moviePage.read().decode('utf-8'), encoding='utf-8' => 문자열 데이터타입을 'utf-8' 인코딩 => 데이터타입: 딕션

#딕션데이터를 활용해보자!!
movie2019_sample = moviePage_json['movieListResult']['movieList']

movie2019_sample[0].keys()

#데이터프레임을 만들어서 거기다 넣기
dfMov_temp = pd.DataFrame()
dfMov = pd.DataFrame()

dfMov_temp = dfMov_temp.append({"movieCd":"","movieNm":"","openDt":"","repGenreNm":"","companyNm":""},ignore_index=True)

for i in range(len(movie2019_sample)):
    dfMov_temp.iat[0,dfMov_temp.columns.get_loc("movieCd")]= movie2019_sample[i]['movieCd']
    dfMov_temp.iat[0,dfMov_temp.columns.get_loc("movieNm")]= movie2019_sample[i]['movieNm']
    dfMov_temp.iat[0,dfMov_temp.columns.get_loc("openDt")]= movie2019_sample[i]['openDt']
    dfMov_temp.iat[0,dfMov_temp.columns.get_loc("repGenreNm")]= movie2019_sample[i]['repGenreNm']
    if movie2019_sample[i]['companys'] == []:
         dfMov_temp.iat[0,dfMov_temp.columns.get_loc("companyNm")] = "None"
    else:
        dfMov_temp.iat[0,dfMov_temp.columns.get_loc("companyNm")]= movie2019_sample[i]['companys'][0]['companyNm']
    dfMov = dfMov.append(dfMov_temp, ignore_index=True)
    
    dfMov_temp.drop(0)
```

```python
dfMov.columns.get_loc('movieCd')
dfMov

#리스트 정리
col_list = ["movieCd","movieNm","repGenreNm","openDt","companyNm"]
dfMov = dfMov[col_list]
dfMov.sort_values(by="movieCd", ascending =True)

dfMov2 = pd.DataFrame()
dfMov2 = dfMov2.append(dfMov.sort_values(by="movieCd", ascending =True), ignore_index=True)

dfMov = dfMov2

#기생충의 정보를 가져온다.
len(dfMov['movieCd'].unique()) == dfMov.shape[0]
dfMov[dfMov['movieNm'] == "기생충"]
```

```python
import pandas as pd
import json
import urllib.request
#import requests

url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=ae65dcfa77374933afc5fd054760188c&openStartDt=2019&openEndDt=2019&itemPerPage=100&curPage="
#curPage
urlPage = urllib.request.urlopen(url)
url_json = json.loads(urlPage.read().decode('utf-8'), encoding="utf-8")
url_json
tot = url_json['movieListResult']['totCnt']
totPage = tot//100+1


#DF선언
movie_DF = pd.DataFrame()
movie_DF_temp = pd.DataFrame()
movie_DF_temp = movie_DF_temp.append({"movieCd":"","movieNm":"","openDt":"","genre":"","companyNm":"","companyCd":""},ignore_index=True)

movList = url_json['movieListResult']['movieList']
#print(movList[0]['movieNm'])
#print(movList[0]['movieCd'])
#print(movList[0]['openDt'])
#totPage만큼 돌리기
for i in range(1):
    URL = url + str(i+1)
    rep = urllib.request.urlopen(URL)
    movie_json = json.loads(rep.read().decode('utf-8'), encoding='utf-8')
    movList = movie_json['movieListResult']['movieList']
    
    for j in range(len(movList)):
    
        movie_DF_temp.iat[0,movie_DF_temp.columns.get_loc("movieNm")] = movList[j]['movieNm']
        movie_DF_temp.iat[0,movie_DF_temp.columns.get_loc("movieCd")] = movList[j]['movieCd']
        movie_DF_temp.iat[0,movie_DF_temp.columns.get_loc("openDt")] = movList[j]["openDt"]
        movie_DF_temp.iat[0,movie_DF_temp.columns.get_loc("genre")] = movList[j]["repGenreNm"]
        if movList[j]["companys"] != []:
            movie_DF_temp.iat[0,movie_DF_temp.columns.get_loc("companyCd")]=movList[j]["companys"][0]["companyCd"]
            movie_DF_temp.iat[0,movie_DF_temp.columns.get_loc("companyNm")]=movList[j]["companys"][0]["companyNm"]
        else:
            movie_DF_temp.iat[0,movie_DF_temp.columns.get_loc("companyCd")]=""
            movie_DF_temp.iat[0,movie_DF_temp.columns.get_loc("companyNm")]=""
        movie_DF = pd.concat([movie_DF,movie_DF_temp],ignore_index=True)
        
        movie_DF_temp.drop(0)
        

movie_DF

    
    
```
