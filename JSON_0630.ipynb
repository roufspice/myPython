## 3. 데이터프레임 만들기

```python
#전체 DF 선언
movieIDDF= pd.DataFrame()

# 1개 영화 정보들에 대해서 처리할 DF 선언
movieIDDF_temp = pd.DataFrame()

movieIDDF_temp = movieIDDF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","openDt":"","directorPeopleNm":""}, ignore_index=True)




num = len(movieIdData["movieListResult"]["movieList"])

for i in range(num):
    movieList = movieIdData["movieListResult"]["movieList"]
    #movieIDDF_temp에 i번째 영화 정보들을 담기
    movieIDDF_temp.at[0,'movieCd'] = movieList[i]['movieCd']
    movieIDDF_temp.at[0,'movieNm'] = movieList[i]['movieNm']
    movieIDDF_temp.at[0,'movieNmEn'] = movieList[i]['movieNmEn']
    movieIDDF_temp.at[0,'openDt'] = movieList[i]['openDt']
    if movieList[i]['directors'] == []:
        movieIDDF_temp.at[0,'directorPeopleNm'] = ""
    else:
        movieIDDF_temp.at[0,'directorPeopleNm'] = movieList[i]['directors'][0]['peopleNm']

    #print(movieIDDF_temp)
    #movieIDDF_temp에 있는 i번째 영화에 대한 정보를 movieIDDF에 추가!
    movieIDDF = movieIDDF.append(movieIDDF_temp,ignore_index = True)
    
    #movieIDDF_temp에 있는 정보들을 초기화!
   
    movieIDDF_temp.drop(0)
    
    #쌤이 하신 방법!(아래코드)
    
    #for j in range(0,len(movieIDDF_temp.columns)):
        #movieIDDF_temp.iat[0,j] = ""

```

+ openDt의 정보를 활용하여 year, month, day의 칼럼의 값을 생성하기
- [힌트:lambda적용, 문자열의 일부 추출하는 방법 생각해보기]

```python
print(movieIDDF['openDt'][0][:4])
print(movieIDDF['openDt'][0][4:6])
print(movieIDDF['openDt'][0][6:])
```

```python
movieIDDF['year'] = movieIDDF['openDt'].apply(lambda x: x[:4])
movieIDDF['month'] = movieIDDF['openDt'].apply(lambda x: x[4:6])
movieIDDF['day'] = movieIDDF['openDt'].apply(lambda x: x[6:])
```

```python
import pandas as pd
import urllib.request
import json

movieURL_2019 = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=ae65dcfa77374933afc5fd054760188c&itemPerPage=100&openStartDt=2019&openEndDt=2019"
 
moviePage_2019 = urllib.request.urlopen(movieURL_2019)

movie2019_Json = json.loads(moviePage_2019.read().decode("utf8"), encoding="utf8")
```

```python
# 데이터프레임만들기
import pandas as pd
movie2019DF = pd.DataFrame()
movie2019DF_temp = pd.DataFrame()

movie2019DF_temp = movie2019DF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","genreNm":"","openDt":""},ignore_index =True)
#movie2019DF_temp


#movie2019DF_temp.columns.get_loc("")
for i in range(len(movie2019_Json['movieListResult']['movieList'])):
    movieIdx = movie2019_Json['movieListResult']['movieList']
    #데이터프레임 한줄씩 집어넣기
    movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('movieCd')] = movieIdx[i]['movieCd']
    movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('movieNm')] = movieIdx[i]['movieNm']
    movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('movieNmEn')] = movieIdx[i]['movieNmEn']
    movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('genreNm')] = movieIdx[i]['repGenreNm']
    movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('openDt')] = movieIdx[i]['openDt']
    
    #데이터프레임 append추가하기
    movie2019DF = movie2019DF.append(movie2019DF_temp, ignore_index=True)
    
    #데이터프레임 초기화하기(temp에 담겨져 있는 정보들을 초기화 하는 부분!)
    movie2019DF_temp.drop(0)

#csvPath = "movie_2019_3.csv"
#movie2019DF.to_csv(csvPath, encoding ='utf-8-sig')
```

# 여러페이지 가져오기

```python
import pandas as pd
import urllib.request
import json

#DataFrame 선언!
movie2019DF = pd.DataFrame()
csv_file_path ="movie2019.csv"

curPage = int(input("가져올 페이지 수를 입력하세요"))
for i in range(curPage):
    
    movie2019_URL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=ae65dcfa77374933afc5fd054760188c&itemPerPage=100&openStartDt=2019&openEndDt=2019"+"&"+str(curPage)
    movie2019_page = urllib.request.urlopen(movie2019_URL)
    #확인중
    print(str(i)+"페이지 완료")
    
    movie2019_Json = json.loads(movie2019_page.read().decode("utf8"), encoding="utf8")
    
    movie2019DF_temp = pd.DataFrame()
    movie2019DF_temp = movie2019DF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","genreNm":"","openDt":""},ignore_index =True)
    
    #forLoop돌리기
    for j in range(len(movie2019_Json['movieListResult']['movieList'])):
        
        movieIdx = movie2019_Json['movieListResult']['movieList']
        #데이터프레임 한줄씩 집어넣기
        movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('movieCd')] = movieIdx[j]['movieCd']
        movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('movieNm')] = movieIdx[j]['movieNm']
        movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('movieNmEn')] = movieIdx[j]['movieNmEn']
        movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('genreNm')] = movieIdx[j]['repGenreNm']
        movie2019DF_temp.iat[0,movie2019DF_temp.columns.get_loc('openDt')] = movieIdx[j]['openDt']
        
        #데이터프레임 append 
        movie2019DF = movie2019DF.append(movie2019DF_temp, ignore_index=True)
        
        #데이터프레임_temp 리셋
        movie2019DF_temp.drop(0)
        

movie2019DF.to_csv(csv_file_path, encoding='utf-8-sig')
```
